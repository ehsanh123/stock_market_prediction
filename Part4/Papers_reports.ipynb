```python
import matplotlib.pyplot as plt

# Data and labels
categories = [
    "Econometric/Statistical Models",
    "Trading Strategies",
    "Machine Learning (ML)",
    "Deep Learning (DL)"
]
sizes = [7, 5, 4, 3]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    sizes,
    labels=categories,
    colors=colors,
    startangle=140
)

plt.title("Studies Using Google Trends in Research", fontsize=16)
plt.axis('equal')  # Ensures the pie is drawn as a circle.

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
methods = [
    "LSTM / Bi-LSTM",
    "ARIMA / GARCH",
    "Random Forest",
    "SVM / SVR",
    "XGBoost",
    "MLP",
    "CNN",
    "Transformer / Attention",
    "Hybrid Models"
]
percentages = [45, 15, 12, 10, 8, 5, 5, 5, 5]

# Plotting
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    percentages, labels=methods, autopct='%1.1f%%', startangle=140,
    colors=plt.cm.Set3.colors
)

# Chart settings
ax.set_title("Stock Prediction Methods Used in Research Papers (20202025)")
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Data for stock indices studied
indices = [
    "S&P 500 (USA)",
    "NASDAQ (USA)",
    "NIFTY/Sensex (India)",
    "Shanghai/CSI 300 (China)",
    "DAX-30 (Germany)",
    "TEDPIX (Iran)",
    "Other Global Indices"
]
percentages = [35, 18, 15, 12, 5, 3, 12]

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.barh(indices, percentages, color='mediumseagreen')
plt.xlabel('Estimated % of Studies')
plt.title('Frequency of Stock Indices Studied in Prediction Research (20202025)')
plt.gca().invert_yaxis()

# Add percentage labels to bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height()/2, f'{width}%', va='center')

plt.tight_layout()
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```
